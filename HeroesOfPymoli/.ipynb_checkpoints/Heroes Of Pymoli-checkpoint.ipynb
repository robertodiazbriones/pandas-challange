```python
#Import Dependecies
import pandas as pd
```

```python
#Path to data
HoP_csv="Resources/HeroesOfPymoli_purchase_data.csv"
#convert data into a Data Frame
HoP_df = pd.read_csv(HoP_csv)
```

```python
#Display Data Frame
HoP_df.head(20)
```

```python
#Display the total number of players

Players_count=len(HoP_df["SN"].unique())
print(Players_count)
#Create Data Frame to display value
Players_count_df=pd.DataFrame({"Total Players":[Players_count]})
Players_count_df
```

```python
#Run basic calculations to obtain number of unique items, average price, etc.
Item_count=len(HoP_df["Item Name"].unique())
Num_Purchases=len(HoP_df["Purchase ID"].value_counts())
Item_Avg_Price=HoP_df["Price"].sum()/Num_Purchases
Total_Revenue=HoP_df["Price"].sum()

#Create Data Frame
Purchasing_Analysis_Total_df=pd.DataFrame({"Number of Unique Items":[Item_count],
                                        "Average Price":[Item_Avg_Price],
                                        "Number of Purchases":[Num_Purchases],
                                        "Total_Revenue":[Total_Revenue]})

#Format data
Purchasing_Analysis_Total_df["Average Price"]=Purchasing_Analysis_Total_df["Average Price"].astype(float).map("${:.2f}".format)
Purchasing_Analysis_Total_df["Total_Revenue"]=Purchasing_Analysis_Total_df["Total_Revenue"].astype(float).map("${:.2f}".format)
Purchasing_Analysis_Total_df.head()
```

```python
# Remove all dublicates
HoP_modified_df=HoP_df.drop_duplicates("SN")

#Group by gender, count by gender in "Count" column
HoP_Gender_df=HoP_modified_df.groupby("Gender").size().reset_index(name='Count')

#Create a new column with the Percentage of Players
HoP_Gender_df["Percentage of Players"]=HoP_Gender_df["Count"]*100/HoP_Gender_df["Count"].sum()

#Format data
HoP_Gender_df["Percentage of Players"]=HoP_Gender_df["Percentage of Players"].astype(float).map("{:.2f}%".format)
HoP_Gender_df
```

```python
HoP_df.head(10)
```

```python
#Find all female Players
Total_Female_df=HoP_df.loc[HoP_df["Gender"] == "Female", :]

#Count Female Players, Sum Column Price for Female players and get Purchase Average
Total_Female=Total_Female_df["Gender"].count()
TotalPur_Female=Total_Female_df["Price"].sum()
AvgPur_Female=TotalPur_Female/Total_Female

#Find all male Players
Total_Male_df=HoP_df.loc[HoP_df["Gender"] == "Male", :]

#Count male Players, Sum Column Price for male players and get Purchase Average
Total_Male=Total_Male_df["Gender"].count()
TotalPur_Male=Total_Male_df["Price"].sum()
AvgPur_Male=TotalPur_Male/Total_Male

#Find all Other / Non-Disclosed Players
Total_Other_df=HoP_df.loc[HoP_df["Gender"] == "Other / Non-Disclosed", :]

#Count male Other / Non-Disclosed Players Players, 
#Sum Column Price for Other / Non-Disclosed Players players and get Purchase Average
Total_Other=Total_Other_df["Gender"].count()
TotalPur_Other=Total_Other_df["Price"].sum()
AvgPur_Other=TotalPur_Other/Total_Other

#Create new df
Purchasing_Gender_df=pd.DataFrame({"Gender":["Female","Male","Other / Non-Disclosed"],
                                  "Purchase Count":[Total_Female, Total_Male, Total_Other],
                                  "Avg Purchase Price":[AvgPur_Female, AvgPur_Male, AvgPur_Other],
                                  "Total Purchase Value":[TotalPur_Female, TotalPur_Male, TotalPur_Other]
                                  })

#Create "Avg Total Purchase per Person" column and populate data with using sata from the "Total Purchase Value", 
#column and the "Count" column from the HoP_Gender_df 
Purchasing_Gender_df["Avg Total Purchase per Person"]=Purchasing_Gender_df["Total Purchase Value"]/HoP_Gender_df["Count"]

#Format data in df
Purchasing_Gender_df["Avg Purchase Price"]=Purchasing_Gender_df["Avg Purchase Price"].astype(float).map("${:.2f}".format)
Purchasing_Gender_df["Total Purchase Value"]=Purchasing_Gender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Purchasing_Gender_df["Avg Total Purchase per Person"]=Purchasing_Gender_df["Avg Total Purchase per Person"].astype(float).map("${:.2f}".format)
Purchasing_Gender_df=Purchasing_Gender_df.groupby("Gender")
Purchasing_Gender_df.max()
```

```python
#Create bins
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
#Create group names
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#create a new data frame
HoP_modified2_df=HoP_modified_df

#Create a new column and categorize the players
HoP_modified2_df["Group Age"]=pd.cut(HoP_modified2_df["Age"], bins, labels=group_names)

#Group by Age and populate total Count column with .agg and rename
Age_Demog_df=(HoP_modified2_df.groupby("Group Age")
            .agg({'Age':'count'})
            .rename(columns={'Age':'Total Count'}))

#Get Percentage of Players with Players_count variable got it before
Age_Demog_df["Percentage of Players"]=Age_Demog_df["Total Count"]*100/Players_count

#Format data
Age_Demog_df["Percentage of Players"]=Age_Demog_df["Percentage of Players"].astype(float).map("{:.2f}%".format)
Age_Demog_df
```

```python
#Create bins
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
#Create group names
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#create a new data frame
HoP_modified3_df=HoP_df

#Create a new column and categorize the players
HoP_modified3_df["Group Age"]=pd.cut(HoP_modified3_df["Age"], bins, labels=group_names)

#Create a new column with same values than price to get average
HoP_modified3_df["Avg"]=HoP_modified3_df["Price"]

#Group by Age and populate data with .agg and rename
Age_Pur_df=(HoP_modified3_df.groupby("Group Age")
            .agg({'Age':'count','Avg':'mean','Price':'sum'})
            .rename(columns={'Age':'Purchase Count', 'Avg':'Average Purchase Price','Price':'Total Purchase Value'}))
            
#Format data    
Age_Pur_df['Avg Total Purchase per Person']=Age_Pur_df['Total Purchase Value']/Age_Demog_df["Total Count"]
Age_Pur_df['Average Purchase Price']=Age_Pur_df['Average Purchase Price'].astype(float).map("${:.2f}".format)
Age_Pur_df['Total Purchase Value']=Age_Pur_df['Total Purchase Value'].astype(float).map("${:.2f}".format)
Age_Pur_df['Avg Total Purchase per Person']=Age_Pur_df['Avg Total Purchase per Person'].astype(float).map("${:.2f}".format)
Age_Pur_df
```

```python
HoP_modified4_df=HoP_df
HoP_modified4_df["Purchase Count"]=HoP_modified4_df["SN"]
HoP_modified4_df["Avg"]=HoP_modified4_df["Price"]
Top_Senders_df=(HoP_modified4_df.groupby("SN").agg({'Purchase Count':'count', 'Avg':'mean','Price':'sum'})
                                             .rename(columns={'Avg':'Average Purchase Price','Price':'Total Purchase Value'}))
Top_Senders2_df=Top_Senders_df.sort_values(by='Purchase Count', ascending= False)
Top_Senders2_df=Top_Senders2_df.iloc[:5]

#Format data    
Top_Senders2_df['Average Purchase Price']=Top_Senders2_df['Average Purchase Price'].astype(float).map("${:.2f}".format)
Top_Senders2_df['Total Purchase Value']=Top_Senders2_df['Total Purchase Value'].astype(float).map("${:.2f}".format)
Top_Senders2_df
```
